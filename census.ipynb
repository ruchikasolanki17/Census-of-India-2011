```python
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv("C:\\Users\\ruchi\\OneDrive\\Desktop\\census of india_2011.csv")
census_data_sorted = dataset.sort_values(by='Population', ascending=False)
plt.figure(figsize=(18, 12))
plt.scatter(census_data_sorted['State'], census_data_sorted['Population'], color='blue')
plt.xticks(rotation=90)
plt.title('Population by State (Census 2011) - Descending Order')
plt.xlabel('State')
plt.ylabel('Population')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv('C:\\Users\\ruchi\\OneDrive\\Desktop\\census of india_2011.csv')
print(dataset.columns)
dataset['Males'] = dataset['Males'].replace(',', '', regex=True).astype(int)
dataset['Females'] = dataset['Females'].replace(',', '', regex=True).astype(int)


male_population = dataset['Males'].sum()
female_population = dataset['Females'].sum()


labels = ['Males', 'Females']
sizes = [male_population, female_population]
colors = ['#1f77b4', '#ff7f0e']  
explode = (0.1, 0)  

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle= 120)
plt.axis('equal') 
plt.title('Population Distribution by Gender (Census of India_2011)')
plt.show()
```
